```python
data = np.genfromtxt('hw1_train.dat')
data
```

Random sampling

```python
def random_sampling(data, x_0):
   #rxy = random.choices(data)
    rxy = random.choice(data)
    rxy = np.insert(rxy, 0, x_0)
    rx = rxy[0:-1]
    ry = rxy[-1]
    return rx, ry
```

PLA

```python
update = []
t = 1000

for i in range(t):
    count = 0
    N_5 = 0
    x_0 = 1

    w = np.zeros(13)
    x, y = random_sampling(data, x_0)
    while N_5 <= (5 * data.shape[0]):
        if  y * np.dot(w, x) < 0 or (y * np.dot(w, x) == 0 and y > 0):
            w = w + (y * x)
            count = count + 1
            N_5 = 0
        else:
            N_5 = N_5 + 1
        x, y = random_sampling(data, x_0)
        
    print(w)
    update.append(count)
    
plt.hist(update, bins = 200)
plt.show()
mdn = np.median(update)
print("Median number of updates is {}".format(mdn))
```
