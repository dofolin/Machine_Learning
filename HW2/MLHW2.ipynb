Import package

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import random
```

Generating data

```python
def generating_data(n):
    #sorting sequence
    x = sorted([random.uniform(-1, 1) for _ in range(n)])
    #signed sequence
    y = [np.sign(np.sign(i)+(2.5 * random.random() - 1.25)) for i in x]
    return x, y
```

Ein & Eout

```python
def Ein(data, s, theta):
    x, y = data
    bad = 0
    #check and count wrong sign
    for i in range(len(y)):
        if (s * np.sign(x[i] - theta)) != y[i]:
            bad += 1
    return bad / len(x)

def Eout(s, theta):
    bad = 0.5 - 0.4 * s + 0.4 * s * abs(theta)
    return bad
```

```python
def decision_stump(data):
    x, y = data


    thetalist = [-1] + [(x[i] + x[i+1]) / 2 for i in range(len(x)-1)]
    h = [(-1, theta) for theta in thetalist] + [(1, theta) for theta in thetalist]

    g = (0, 0)
    einf = float('inf')

    for s, theta in h:
        ein = Ein(data, s,theta)
        if ein <= einf:
            if ein == einf:
                if s * theta < g[0] * g[1]:
                    g = (s, theta)
            else:
                einf = ein
                g = (s, theta)
    return g

    return g
```

10

```python
#create lists for compute
einlist = []
eoutlist = []
#repeat 2000 times
for i in range(2000):
    data = generating_data(32)
    #dataset size: 32
    g = decision_stump(data)
    #calculate g by decision stump
    ein = Ein(data, g[0], g[1])
    einlist.append(ein)
    #calculate eout
    eout = Eout(g[0], g[1])
    eoutlist.append(eout)

median = np.median(np.array(eoutlist) - np.array(einlist))
#calculate median

plt.scatter(einlist, eoutlist)
plt.xlabel('Ein')
plt.ylabel('Eout')
#plot scatter
print('the median of Eout(g)-Ein(g):', median)
```
